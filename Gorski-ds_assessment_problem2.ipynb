```python
import pandas as pd

au_train = pd.read_csv("au_train.csv")
au_test = pd.read_csv("au_test.csv")
dataset = pd.concat([au_train, au_test])
```

```python
dataset.head(3)
```

```python
# Checking for missing values.

dataset.isnull().sum()
```

```python
dataset.isin([" ?"]).sum()
```

```python
# Removing missing '?' values.
dataset = dataset.replace([" ?","NaN"])
```

```python
dataset = dataset.dropna()
```

```python
dataset.isin([" ?"]).sum()
dataset.head(3)
```
